## 3.4 Augment data with additional columns

**Note: In US have city same name in different states**

__Lưu ý: Ở Mỹ có thành phố có cùng tên ở các tiểu bang khác nhau__

#### Add city column

```python
# Create function take city from "Purchase Address" columns
def get_address(address):
    city = address.split(",")[1].strip(" ")
    state = address.split(",")[2].split(" ")[1]
    result = city + " " + "(" + state + ")"
    return result

#strip(): bỏ khoảng trắng
```

```python
# Applying a function to "Purchase Address" columns then create new columns "City"
all_data['City'] = all_data['Purchase Address'].apply(get_address)

# random sample of rows from the data frame
all_data.head(5)
```

# 4. Data Wrangling

## 4.1 **Question 1:** What was the best month for sales? How much was earned that month?

__Tháng tốt nhất để bán hàng là gì? Tháng đó kiếm được bao nhiêu?__


#### Add month column

```python
# Create "Month" columns from extract data in "Order Date"
all_data['Month'] = all_data['Order Date'].dt.month
all_data.sample(5)
```

#### Add Sales column

```python
# Create "Sales" columns from "Quantity Ordered" * "Price Each"
all_data['Sale'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
# Group month and sum "Sales" column
total_month_df = all_data.groupby(by='Month').agg({'Sale':sum})
total_month_df
```

```python
max_sales = np.max(total_month_df['Sale'])
max_sales
```

```python
min_sales = np.min(total_month_df['Sale'])
min_sales
```

```python
highest_df = total_month_df[total_month_df['Sale'] == max_sales]
highest_df
```

```python
highest_month = highest_df.idxmax()['Sale']
highest_month
```

```python
print("The month that the best month for sales is {0} and it earned {1} for that month".format(highest_month,max_sales))
```

```python
# Plot bar chart sales of 12 months with pandas plot
total_month_df.plot(kind='bar', figsize=(16, 9), xlabel='Month number', ylabel='Sales in USD ($)');
```

```python
max_city = np.max(total_city_df['Sale'])
max_city
```

```python
highest_city = highest_city_df.idxmax()['Sale']
highest_city
```

```python
print("The city had the highest number of sales is {0} for {1} $.".format(highest_city,max_city))
```

## 4.4 **Question 4:** What product was the best Year for sales? How much was earned that Year?

__Sản phẩm nào là năm bán chạy nhất? Năm đó kiếm được bao nhiêu?__

```python
product_group_df = all_data.groupby(['Product']).agg({"Quantity Ordered":"sum","Sale":"sum","Price Each": "mean"})
product_group_df
```

### The product was the best Year for sales

```python
max_product_sales = np.max(product_group_df["Quantity Ordered"])
max_product_sales
```

```python
highest_sales_product_df = product_group_df[product_group_df["Quantity Ordered"] == max_product_sales]
highest_sales_product_df
```

```python
highest_sales_product= highest_sales_product_df.index[0]
highest_sales_product
```

### The product was the best Year for earn

```python
max_earn_product = np.max(product_group_df["Sale"])
max_earn_product
```

```python
highest_earn_product_df = product_group_df[product_group_df["Sale"] == max_earn_product]
highest_earn_product_df

highest_earn_product= highest_earn_product_df.index[0]
highest_earn_product
```

```python
print("The product was the best Year for sales is {0} and it earn {1} $.".format(highest_sales_product,max_product_sales))
print("The product was the best Year for earn is {0} and it earn {1} $.".format(highest_earn_product,max_earn_product))
```

```python
# Plot with matplotlib
fig, ax1 = plt.subplots(figsize=(16,9))


ax1.bar(product_group_df.index, product_group_df['Quantity Ordered'], color='g')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax1.set_xticklabels(product_group_df.index , rotation='vertical', size=10)

ax2 = ax1.twinx()
ax2.plot(product_group_df.index, product_group_df['Sale'], color='b',marker = 'o')
ax2.set_ylabel('Earn ($)', color='b')
ax2.set_xticklabels( product_group_df.index,rotation='vertical', size=10);
fig.show()
```

## 4.5 **Question 5:** What is the monthly sales of each item? Which products have the most seasonal sales?

```python
# Pivot table with "Month" and products
df2 = pd.pivot_table(data = all_data, index="Product", columns = "Month", values = "Sale", aggfunc = "sum")
df2 
```
